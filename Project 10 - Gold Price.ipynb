```python
corr['GLD']
```

```python
sns.distplot(gold_data['GLD'],color='Green')
```

```python
# Splitting

X = gold_data.drop(columns = ['Date','GLD'],axis=1)
Y = gold_data['GLD']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(X_train,Y_train)
```

```python
predict_train = regressor.predict(X_train)
r2_train = metrics.r2_score(predict_train, Y_train)
r2_train
```

```python
predict_test = regressor.predict(X_test)
r2_test = metrics.r2_score(predict_test, Y_test)
r2_test
```

```python
# comparing the actual and predicted values

Y_test = list(Y_test)
```

```python
plt.plot(Y_test, color='Blue', label = ' Actual Value')
plt.plot(predict_test, color = 'Green', label = 'Predicted Value')
plt.xlabel('No. of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
