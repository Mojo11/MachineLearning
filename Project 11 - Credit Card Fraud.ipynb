```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_data = pd.read_csv("/content/creditcard.csv")
```

```python
credit_data.shape
```

```python
credit_data.head()
```

```python
credit_data.isnull().sum()
```

```python
credit_data.dropna(inplace=True)
```

```python
credit_data.isnull().sum()
```

```python
credit_data['Class'].value_counts()
```

```python
non_fraud = credit_data[credit_data.Class == 0.0]
fraud = credit_data[credit_data.Class == 1.0]
```

```python
non_fraud = non_fraud.sample(n=368)
```

```python
non_fraud.shape
```

```python
fraud.shape
```

```python
new_credit = pd.concat([non_fraud, fraud], axis=0)
```

```python
new_credit.shape
```

```python
X = new_credit.drop('Class', axis=1)
Y = new_credit['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state =2)
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train, Y_train)
```

```python
predict_train = classifier.predict(X_train)
accuracy_train = accuracy_score(predict_train, Y_train)
accuracy_train
```

```python
predict_test = classifier.predict(X_test)
accuracy_test = accuracy_score(predict_test, Y_test)
accuracy_test
```
