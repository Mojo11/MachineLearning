```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
X = heart_data.drop(columns = 'target', axis = 1)
Y = heart_data['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
predict_train = model.predict(X_train)
accuracy_train = accuracy_score(Y_train, predict_train)
accuracy_train
```

```python
predict_test = model.predict(X_test)
accuracy_test = accuracy_score(Y_test, predict_test)
accuracy_test
```

# K Fold Cross Validation

Comparing the performance of the models.

```python
models = [LogisticRegression(max_iter = 1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_models_train_test():

  for model in models:
    model.fit(X_train, Y_train)
    test_data_predict = model.predict(X_test)
    accuracy = accuracy_score(Y_test, test_data_predict)

    print('Accuracy of the ', model, ' = ', accuracy)
```

```python
compare_models_train_test()
```

#Cross Validation Score (cross_val_score)

```python
# Logistic Regression = lr

cv_score_lr = cross_val_score(LogisticRegression(max_iter=1000), X, Y, cv = 5)

print(cv_score_lr)

accuracy_lr = np.sum(cv_score_lr) / (len(cv_score_lr))
accuracy_lr = round(accuracy_lr*100,2)

accuracy_lr
```

```python
# SVM

cv_score_svm = cross_val_score(SVC(kernel = 'linear'), X, Y, cv = 5)

print(cv_score_svm)

accuracy_svm = np.sum(cv_score_svm) / (len(cv_score_svm))
accuracy_svm = round(accuracy_svm*100,2)

accuracy_svm
```

```python
# KNeighbors

cv_score_knn = cross_val_score(KNeighborsClassifier(), X, Y, cv = 5)

print(cv_score_knn)

accuracy_knn = np.sum(cv_score_knn) / (len(cv_score_knn))
accuracy_knn = round(accuracy_knn*100,2)

accuracy_knn
```

```python
# Random Forest

cv_score_rfc = cross_val_score(RandomForestClassifier(), X, Y, cv = 5)

print(cv_score_rfc)

accuracy_rfc = np.sum(cv_score_rfc) / (len(cv_score_rfc))
accuracy_rfc = round(accuracy_rfc*100,2)

accuracy_rfc
```

creating a function to evaluate the cross validation on all the models

```python
def compare_models_cross_var():

  for model in models:

    cv_score = cross_val_score(model, X, Y, cv=5)
    mean_accuracy = np.sum(cv_score)/len(cv_score)

    mean_accuracy = round(mean_accuracy*100, 2)

    print('Cross Val accuracies for the ', model, ' = ', cv_score)
    print(' Accuracy % of the, ', model, mean_accuracy)
```

```python
compare_models_cross_var()
```
